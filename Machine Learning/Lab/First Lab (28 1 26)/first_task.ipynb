```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load a dataset from the internet
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)

print (df)
```

```python
# One independent variable (X) and one dependent variable (Y)
x = df["total_bill"].values
y = df["tip"].values

print("X:", x)
print("Y:", y)
```

```python
#plotting
plt.scatter(x,y)
plt.xlabel("Total Bill")
plt.ylabel("Tips")
plt.title("Total Bill vs Tips")
plt.show()

#graphs shows Linear Relationship
```

```python
#mean
x_mean = np.mean(x)
y_mean = np.mean(y)

print ("Mean of X: ", x_mean)
print ("Mean of Y:", y_mean)
```

```python
#slope
numerator = np.sum((x-x_mean) * (y-y_mean))
denominator = np.sum ((x-x_mean)**2)

m = numerator/denominator
print ("Slope (a):", m)
```

```python
#intercept
c = y_mean - m *x_mean
print (c)
```

```python
print (f"Regression Equation y = {m:.2f}x + {c:.2f}")
```

```python
y_pred = m * x + c
print ("Predicted Values:", y_pred)
```

```python
plt.scatter(x,y, label= "Actual Data")
plt.plot(x, y_pred, color = 'red', label = "Regrerssion Line")
plt.xlabel("Hours Studies")
plt.ylabel("Marks Obtained")
plt.legend()
plt.show()
```

```python
errors = y - y_pred
print (errors)
```

```python
mse = np.mean(errors ** 2)
print ("Mean Squared Error:", mse)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import mean_square_error

X = x.reshape(-1, 1)
model = LinearRegression()
model.fit(X, y)

# line for plotting
x_line = np.linspace(x.min(), x.max(), 100)
y_line = model.predict(x_line.reshape(-1, 1))

plt.scatter(x, y, label="Data")
plt.plot(x_line, y_line, color="red", label="Linear fit")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression Fit")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression

X = x.reshape(-1, 1)
model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)
mse = mean_squared_error(y, y_pred)
print("MSE:", mse)
```
